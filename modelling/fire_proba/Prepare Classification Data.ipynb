```python
from random import randrange

import numpy as np
import pandas as pd

import requests
```

```python
def get_features(lat, lng):
    url = "http://api.apixu.com/v1/current.json?key=ad23cdc4a1404d34984174745182010&q=" + str(lat) + "," + str(lng)
    r = requests.get(url).json()['current']
    data = {
        "temp": r['temp_c'],
        "RH": r['humidity'],
        "wind": r['wind_kph'],
        "rain": r['precip_mm']
    }
    return data

def get_random_cords():
    return randrange(-90, 90), randrange(-180, 180)
```

```python
fires = pd.read_csv("../fires.csv")
```

```python
fires = fires[fires.columns[:5]]
```

## Add Meteorological Data from Random Cords

```python
for _ in range(600):
    lat, lng = get_random_cords()
    fires = fires.append(
        pd.Series(get_features(lat, lng)),
        ignore_index=True
    )
```

```python
fires = fires.fillna(0)
fires["area"] = [0 if fires["area"][i] == 0 else 1 for i in range(fires.shape[0])]
```

```python
fires.columns = ["temp", "RH", "wind", "rain", "fire"]
```
